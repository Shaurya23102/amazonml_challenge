```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords

# Download stopwords if not already
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
df = pd.read_csv('/content/train.csv')

# Load your dataset (assuming 'df' is defined and has 'catalog_content')
# df = pd.read_csv("your_file.csv")

# --- MODIFIED FUNCTION ---
def clean_text(text):
    # Lowercase
    text = str(text).lower()

    # MODIFICATION: Keep 'a-z', '0-9', and whitespace '\s'
    # The pattern r'[^a-z0-9\s]' now removes anything that IS NOT a letter, number, or space.
    text = re.sub(r'[^a-z0-9\s]', '', text)

    # Remove stopwords
    words = [word for word in text.split() if word not in stop_words]

    # Join and remove extra spaces that might be left from the re.sub
    return ' '.join(words)

def remove_word(text, word_to_remove):
    # \b ensures only whole words are matched
    pattern = r'\b' + re.escape(word_to_remove.lower()) + r'\b'
    text = re.sub(pattern, '', str(text).lower())
    # Remove extra spaces
    return re.sub(r'\s+', ' ', text).strip()

# Apply cleaning
df['catalog_content'] = df['catalog_content'].apply(clean_text)

word = "item name"

# Apply word removal
df['catalog_content'] = df['catalog_content'].apply(lambda x: remove_word(x, word))
df
```

```python
import torch
import torch.nn as nn
from transformers import AutoModel

# Custom SMAPE Loss function
def smape_loss(preds: torch.Tensor, actuals: torch.Tensor) -> torch.Tensor:
    """
    Calculates the Symmetric Mean Absolute Percentage Error (SMAPE).
    Formula: (1/n) * Σ |preds - actuals| / ((|actuals| + |preds|)/2)
    """
    numerator = torch.abs(preds - actuals)
    denominator = (torch.abs(actuals) + torch.abs(preds)) / 2
    # Add a small epsilon to the denominator to prevent division by zero,
    # though for price data, this is less common.
    epsilon = 1e-8

    # Calculate the percentage error for each element
    percentage_error = numerator / (denominator + epsilon)

    # The mean of the percentage errors is the final SMAPE
    return torch.mean(percentage_error)

class PriceRegressor(nn.Module):
    def __init__(self, model_name):
        super().__init__()
        self.encoder = AutoModel.from_pretrained(model_name)
        # Assuming the output dimension is 1 for a single price prediction
        self.regressor = nn.Linear(self.encoder.config.hidden_size, 1)

    def forward(self, input_ids, attention_mask, labels=None):
        # Pass inputs through the pre-trained model encoder
        outputs = self.encoder(input_ids=input_ids, attention_mask=attention_mask)

        # Extract the hidden state for the [CLS] token (first token)
        pooled = outputs.last_hidden_state[:, 0, :]

        # Pass the pooled representation through the linear regressor
        # Squeeze the last dimension to make it a (batch_size,) tensor
        preds = self.regressor(pooled).squeeze(-1)

        loss = None
        if labels is not None:
            # Use the custom SMAPE loss function
            loss = smape_loss(preds, labels)

        return {"loss": loss, "preds": preds}
```

```python
import torch
import torch.nn as nn
from transformers import AutoModel, AutoTokenizer
from tqdm import tqdm
from datasets import Dataset
from torch.utils.data import DataLoader
import pandas as pd
from sklearn.model_selection import train_test_split

# --- 1. Prepare Data ---
df = df.dropna(subset=["catalog_content", "price"])
df["catalog_content"] = df["catalog_content"].astype(str)
df["price"] = df["price"].astype(float)

train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize(batch):
    return tokenizer(
        batch["catalog_content"],
        padding="max_length",
        truncation=True,
        max_length=400
    )

train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)

train_dataset = train_dataset.map(tokenize, batched=True)
val_dataset = val_dataset.map(tokenize, batched=True)

train_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "price"])
val_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "price"])

BATCH_SIZE = 16
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)


# --- 2. SMAPE Loss ---
def smape_loss(preds: torch.Tensor, actuals: torch.Tensor) -> torch.Tensor:
    numerator = torch.abs(preds - actuals)
    denominator = (torch.abs(actuals) + torch.abs(preds)) / 2
    epsilon = 1e-8
    return torch.mean(numerator / (denominator + epsilon))


# --- 3. Model Definition (with Dropout) ---
class PriceRegressor(nn.Module):
    def __init__(self, model_name):
        super().__init__()
        self.encoder = AutoModel.from_pretrained(model_name)
        self.dropout = nn.Dropout(p=0.3)  # Added dropout layer
        self.regressor = nn.Linear(self.encoder.config.hidden_size, 1)

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.encoder(input_ids=input_ids, attention_mask=attention_mask)
        pooled = outputs.last_hidden_state[:, 0, :]
        pooled = self.dropout(pooled)  # Apply dropout
        preds = self.regressor(pooled).squeeze(-1)
        loss = None
        if labels is not None:
            loss = smape_loss(preds, labels)
        return {"loss": loss, "preds": preds}


# --- 4. Training Setup ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = PriceRegressor(model_name).to(device)

# L2 regularization via weight decay
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5, weight_decay=0.01)

scaler = torch.cuda.amp.GradScaler() if torch.cuda.is_available() else None


# --- 5. Training Loop ---
print(f"Starting training on device: {device}")
for epoch in range(7):
    model.train()
    total_loss = 0.0

    for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}"):
        optimizer.zero_grad(set_to_none=True)

        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["price"].float().to(device)

        if device.type == "cuda":
            with torch.cuda.amp.autocast():
                outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
                loss = outputs["loss"]
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs["loss"]
            loss.backward()
            optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    print(f"Epoch {epoch+1} | SMAPE Loss: {avg_loss:.4f}")
```

```python
import torch
import pandas as pd
from transformers import AutoModel, AutoTokenizer
from torch.utils.data import DataLoader
from datasets import Dataset

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Model and tokenizer
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

class PriceRegressor(torch.nn.Module):
    def __init__(self, model_name):
        super().__init__()
        from transformers import AutoModel
        self.encoder = AutoModel.from_pretrained(model_name)
        self.dropout = torch.nn.Dropout(p=0.3)
        self.regressor = torch.nn.Linear(self.encoder.config.hidden_size, 1)

    def forward(self, input_ids, attention_mask):
        outputs = self.encoder(input_ids=input_ids, attention_mask=attention_mask)
        pooled = outputs.last_hidden_state[:, 0, :]
        pooled = self.dropout(pooled)
        preds = self.regressor(pooled).squeeze(-1)
        return {"preds": preds}

# Initialize and load weights
model = PriceRegressor(model_name).to(device)
model.load_state_dict(torch.load("/content/price_regressor.pt", map_location=device))
model.eval()

# --- Inference ---
test_df = pd.read_csv("/content/test.csv")
test_df["catalog_content"] = test_df["catalog_content"].astype(str)

# Tokenize in batches to save RAM
BATCH_SIZE = 16
preds_list = []

for i in range(0, len(test_df), BATCH_SIZE):
    batch_texts = test_df["catalog_content"].iloc[i:i+BATCH_SIZE].tolist()
    encodings = tokenizer(
        batch_texts,
        padding=True,
        truncation=True,
        max_length=400,
        return_tensors="pt"
    )
    input_ids = encodings["input_ids"].to(device)
    attention_mask = encodings["attention_mask"].to(device)

    with torch.no_grad():
        preds = model(input_ids=input_ids, attention_mask=attention_mask)["preds"]
        preds_list.extend(preds.cpu().numpy())

# Save predictions
test_df["predicted_price"] = preds_list
test_df.to_csv("/content/test_with_predictions.csv", index=False)
print("Predictions saved to /content/test_with_predictions.csv")
```
